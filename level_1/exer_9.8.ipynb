```python
#获取股票代码列表
def urlTolist(url):
    allCodeList = []
    html = urllib.request.urlopen(url).read()
    html = html.decode('utf8')
    s = r'<li><a href="http://quote.eastmoney.com/\S\S(.*?).html" target="_blank">'
    print(html)
    pat = re.compile(s)
    code = pat.findall(html)
    print(code)
    for item in code:
        if item[0]=='6' or item[0]=='3' or item[0]=='0':
            allCodeList.append(item)
    return allCodeList
    
```
